Loss Function

```python
def loss_function(b, m , points):
    total_error = 0
    for _, row in points.iterrows():
        x = row['time_study']
        y = row['Marks']
        total_error += ((b + m * x)- y) **2
    return total_error / (2.0 * len(points))
```

```python
def gradient_descent(b, m, points, lr, epochs):
  
    x = points['time_study'].to_numpy()
    y = points['Marks'].to_numpy()
    history = []
    

    for _ in range(epochs):
        err = (b + m * x) - y
        db = err.mean()          # ∂J/∂b (half-MSE)
        dm = (err * x).mean()    # ∂J/∂m
        temp_b = b - lr * db     # simultaneous update
        temp_m = m - lr * dm
        b, m = temp_b, temp_m
        
    return b, m
    
```

```python
b = 0
m = 0
lr = 0.0001
epochs = 1000

b , m = gradient_descent(m, b, df, lr, epochs)

print(b,m)
```
